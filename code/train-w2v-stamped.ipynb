```python
output_dir = "~/Documents/GitHub/CrisisBureaucracy/results/w2v-models"
```

```python
list_bins = [[x,x+1] for x in range(1957,1984)]


for window in [12,16,24]:
    for count,time_ in enumerate(list_bins[1:]):

        min_count = 50
        workers = 8
        iter = 12
        size = 100

        if count == 0:
            data = data_loader.subset(data_version="lemmatized_pm",start_date=time_[0],end_date=time_[1] + 1,words=[],exact_match=True,preprocess=True).reset_index(drop=True)
            data['text_lemmatized'] = [i.replace("bureaucracy",f"bureaucracy_{data['speech_member_party'][c].lower()}") for c,i in enumerate(data['text_lemmatized'])]
            texts = list(data['text_lemmatized'])
            texts = [t.split(' ') for t in texts]
            print(time_[0],'-',time_[1],len(texts))
            model = gensim.models.Word2Vec(texts, min_count=min_count, workers=workers, iter=iter, size = size, window = window)
            model.wv.save_word2vec_format(os.path.join(f"{output_dir}/{time_[0]}-{time_[1]}-w{window}-model.bin"), binary=True)
        if count != 0:
            data = data_loader.subset(data_version="lemmatized_pm",start_date=time_[0],end_date=time_[1] + 1,words=[],exact_match=True,preprocess=True).reset_index(drop=True)
            data['text_lemmatized'] = [i.replace("bureaucracy",f"bureaucracy_{data['speech_member_party'][c].lower()}") for c,i in enumerate(data['text_lemmatized'])]
            texts = list(data['text_lemmatized'])
            texts = [t.split(' ') for t in texts]
            model.build_vocab(texts, update=True)
            model.train(texts, total_examples = model.corpus_count, start_alpha = model.alpha, end_alpha = model.min_alpha, epochs = model.iter)
            model.wv.save_word2vec_format(os.path.join(f"{output_dir}/{time_[0]}-{time_[1]}-w{window}-model.bin"), binary=True)
            print(time_[0],'-',time_[1],len(texts))
```
